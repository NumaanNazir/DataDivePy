```python
import matplotlib.pyplot as plt
plt.plot([1,3,5,7,9,11],[2,4,6,8,10,5] , "r" ,label="Men")
plt.plot([1,3,5,7,9,11],[2,4,6,8,10,2] ,"g",label="Women" )
plt.legend()
plt.grid()
plt.title("DATA")
plt.xlabel("MoodSwings")
plt.ylabel("Frequency")
plt.show()
```
